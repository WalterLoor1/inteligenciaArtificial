```python
import pandas as pd

# Cargar el archivo CSV
url = "/content/daily_weather.csv"
df = pd.read_csv(url)

# Contar los valores nulos en cada columna
valores_nulos_por_columna = df.isnull().sum()
print("Valores nulos por columna antes de eliminar filas con valores nulos:")
print(valores_nulos_por_columna)

# Eliminar las filas con valores nulos
df = df.dropna()
df = df.drop(columns=['number'])

# Verificar que no hay valores nulos después de la eliminación
valores_nulos_por_columna_despues = df.isnull().sum()
print("Valores nulos por columna después de eliminar filas con valores nulos:")
print(valores_nulos_por_columna_despues)

# Agregar una nueva columna 'Clima' basada en la columna 'relative_humidity_9am'
df['Clima'] = df['relative_humidity_9am'].apply(lambda x: 1 if x > 36 else 0)

# Mostrar las primeras filas del DataFrame actualizado
print(df.head(3))

# Opcional: guardar el DataFrame actualizado en un nuevo archivo CSV
df.to_csv("/content/daily_weather_updated.csv", index=False)
df.head(3)
```
